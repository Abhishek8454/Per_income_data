```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
plt.scatter(df["year"],df["per capita income (US$)"])
plt.xlabel("Year")
plt.ylabel("Per capita income (US$)e")
plt.show()
```

```python
X= df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test.iloc[[0]])
```

```python
df.columns
```

```python
plt.scatter(df["year"],df["per capita income (US$)"],color = "black")
plt.plot(X_train,lr.predict(X_train),color = "red")
plt.xlabel("Year")
plt.ylabel("Per capita income (US$)e")
plt.show()
```

```python
m = lr.coef_
b = lr.intercept_
```

```python
x = float(input("Year"))
y = m*x+b
print("Your package will be around: ",y)
```
